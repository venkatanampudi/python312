# Control flow in Python 

#### We often only want to certain code to execute when a particular condition has been met. 
#### For ex: **if** my dog is hungry __(some condition)__, then I will feed the dog __(some action)__
#### To control this flow of logic we can use __if,elif,else__ key words
#### Control flow syntax makes use of colons and indentation (whitespace)


if some_condition:
    execute code 
elif some_other_condition:
    do something different
else:
    do something else

```python
if True:
    print('Its TRUE')
```

```python
if False:
    print('Its FALSE')
```

```python
if 3>2:
    print('Its True')
```

```python
hungry = True

if True:
    print('Feed Me')
```

```python
hungry = False

if True:
    print('Feed Me')
```

```python
hungry = True

if True:
    print('Feed Me')
else:
    print('Im not hungry')
```

```python
loc = 'Bank'

if loc == 'Auto Shop':
    print("Cars are cool")
else:
    print("I do not know much")
```

```python
loc = 'Bank'

if loc == 'Auto Shop':
    print('Cars are cool')
elif loc == 'Bank':
    print('Money is cool')
else:
    print('I do not know much')
```

```python
# we can stack on many elfi conditions if needed
loc = 'Game'

if loc == 'Auto Shop':
    print('Cars are cool')
elif loc == 'Bank':
    print('Money is cool')
elif loc == 'Store':
    print('Welcome to the store')
else:
    print('I do not know much')
```

```python
name = 'Frankie'
if name == 'Frankie':
    print('Hello Frankie')
elif name == 'Arjun':
    print('Hello Arjun')
else:
    print('What is your name?')
```
