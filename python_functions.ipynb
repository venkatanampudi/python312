# Functions 
#### Functions allow us to create blocks of code that can be easily executed many times, without needing to constantly rewrite the entire block of code.
##### function definition by **def** keyword 
##### __def__ name_of_function():
##### function name is in snake case (all lowercase with underscores between words)
##### Functions can accept arguments to be passed by the user
##### Typically we use the __return__ keyword to send back the result of the funciton, insead of just printing
##### __return__ allows us to assign the output of the function to a new variable
#### Syntax
def name_of_function(name):
    '''
    Docstring explains function.
    '''
    print("Hello" + name)

>> name_of_function("Venkat") ---> to call the function to execute
>>  Hello Venkat
#### Notice everything inside the function is indented.
def add_function(num1,num2):
    return num1+ num2
>> result = add_function(1,2)
>> print(result)
>> 3

```python
# RETURN TRUE IF ANY NUMBER IS EVEN INSIDE A LIST
def check_even_list(num_list):
    
    for number in num_list:
       if number % 2 == 0:
           return True
       else:
           pass
# this function loops thru num_list until it finds the even and then return True. Once it find the even number, func won't run any more    
```

```python
check_even_list([1,3,5]) # func does not return anything if no even number in the list
```

```python
check_even_list([100,200,999])
```

```python
check_even_list([1,1,1,10])
```

```python
def check_even_list(num_list):
    
    for number in num_list:
       if number % 2 == 0:
           return True
       else:
           pass    # THIS pass is just a place holder and dont do anything
    return False   # THIS IS CORRECT
```

```python
check_even_list([3,3,2]) 
```

```python
check_even_list([1,1,1]) 
```

```python
check_even_list([9,9,100]) 
```

```python
# For example - to return all the even numbers in a list called "even_numbers_list" 
def check_even_list(num_list):
    # placeholder variables 
    even_numbers_list = []
    
    for number in num_list:
       if number % 2 == 0:
           even_numbers_list.append(number)
       else:
           pass    # THIS pass is just a placeholder and do not perform anything if number is not "even"
    return even_numbers_list
```

```python
check_even_list([1,2,3,4,5,6,7]) 
```

```python
check_even_list([9,9,9,9])
```

```python
check_even_list([0,1,150,99,1000])
```
