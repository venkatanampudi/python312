```python
# Using keys and indexing, grab the 'Hello' from the following dict
d1 = {'simple_key':'Hello'}
# grab 'Hello'
d1['simple_key']
```

```python
d2 = {'k1':{'k2':'Hello'}}
# grab 'Hello'
d2['k1']['k2']
```

```python
# grab 'Hello' from this dict
d3 = {'k1':[{'nest_key':['This is deep',['Hello']]}]}
d3['k1'][0]['nest_key'][1][0]
```

```python
# grab 'Hello' 
d4 = {'k1':[1,2,{'k2':['This is tricky',{'tough':[1,2,['Hello']]}]}]}
d4['k1'][2]['k2'][1]['tough'][2][0]
```
