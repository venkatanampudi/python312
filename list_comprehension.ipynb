# List comprehensions

```python
mystring = 'Hello' 
```

```python
# 1. create a empty list 
# 2. iterate thru string object and append to empty list
mylist = []

for letter in mystring:
    mylist.append(letter)
```

```python
mylist
```

```python
mylist = [letter for letter in mystring]
```

```python
mylist
```

```python
celcius = [0,1,2,4,10]

fahrenheit = [((9/5)*temp + 32) for temp in celcius]
```

```python
fahrenheit
```

```python
# converting the above with for loop from flatten version
fahrenheit = []

for temp in celcius:
    fahrenheit.append((9/5)*temp + 32)
```

```python
fahrenheit
```

```python
c = [10,20,30,40,50,66.5]

f = [((9/5)*c+32) for c in c]
```

```python
# Using if else in list comprehension. Observe order changed. Insert the condition after first variable and before "for" keyword
results = [num if num%2==0 else 'ODD NUMBER' for num in range(0,11)]
```

```python
results
```

```python
results = ['EVEN' if num%2==0 else 'ODD' for num in range(0,11)]
```

```python
# Nested for loop in list comprehensions

mylist = []

for x in [2,4,6]:
    for y in [1,10,100]:
        mylist.append(x*y)

# take first number and multiply with all second list
```

```python
mylist
```
