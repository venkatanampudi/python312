```python
class Book():

    def __init__(self,title,author,pages):

        self.title = title
        self.author = author
        self.pages = pages 
    # __str__ return the string representation of Book class with whatever defined in this __str__ class. Here, self.title and self.author
    def __str__(self): 
        return f"{self.title} by {self.author}"

    # __len__ return the length of the Book class with defined definition.Here, self.pages
    def __len__(self):
        return self.pages
    # __del__ --> print something up on deleting the object
    def __del__(self):
        print("A book object has been deleted")
```

```python
b = Book('Python Rocks','Jose',200)
```

```python
str(b)
```

```python
print(b)
```
