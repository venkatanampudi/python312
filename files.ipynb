# File locations 
##### If you want to open the files at another location on your computer, simply pass in the entire file path.
##### For Windows : Need to use __double__ \ so Python doesn't treat the second \ as an escape char. 
##### myfile=open("C:\\Users\\UserName\\Folder\\file.txt") (or) use __pwd__ to get the path of the file
##### For Mac OS & Linux: Use slashes in opposite direction(forward slash). myfile=open("/Users/UserName/Folder/file.txt")

# File Reading, Writing, Appending Modes 
* mode='r' is read only
* mode='w' is write only (will overwrite files or create new file if that does exist)
* mode='a' is append only (will add on to files)
* mode='r+' is reading and writing
* mode='w+' is writing and reading (overwrites the existing files or creates a new file)

```python
%%writefile my_new_file.txt
ONE on FIRST
TWO on SECOND 
THREE on THIRD
```

```python
with open('my_new_file.txt',mode='r') as f:
    print(f.read())
```

```python
with open('my_new_file.txt',mode='a') as f:
    f.write('FOUR on FOURTH')
```

```python
with open('my_new_file.txt',mode='r') as f:
    print(f.read())
```
