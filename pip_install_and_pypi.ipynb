### Pip Install & PyPi 
#### PyPi is a repository for open-source 3rd party Python packages
#### It is similar to RubGems in the Ruby world, PHP's Packagist, CPAN for Perl, NPM for Node.js, and CRAN for R language
#### So far we've really only used libraries that come internally with Python 
#### There are many other libraries available that people have open-sourced and shared on PyPi
#### We can use __pip install__ at the commend line to install these packages
#### pip install package-name
#### always install first then import
