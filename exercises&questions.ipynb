```python
def my_decorator(func):
    def wrapper():
        print("Before the function runs")
        func()                       # call the original function
        print("After the function runs")
    return wrapper
```

```python
@my_decorator
def say_hello():
    print('Hello')
```

```python
say_hello()
```
