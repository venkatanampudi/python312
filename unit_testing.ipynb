### unittest

```python
%%writefile cap.py

def title(text):
    '''
    Input a string 
    Output a capitalized string
    '''
    # return text.capitalize() # capitalize just capitalizes the first letter of the sentence. Use title() to capitalize every word in a sentence
    return text.title()
```

```python
%%writefile test_cap.py

import unittest
import cap #importing cap.py module

class TestCap(unittest.TestCase):

    def test_one_word(self):
        text = 'python'
        result = cap.title(text) #using module.method of cap.py
        self.assertEqual(result,'Python')

    def test_multiple_words(self):
        text = 'Monty Python'
        result = cap.title(text)
        self.assertEqual(result,'Monty Python')

if __name__ == '__main__':
    unittest.main()
```

```python
# RUN UNITTEST IN PYTHON
!python test_cap.py
```

```python
import unittest

def add(a, b): # THIS FUNCTION TO TEST
    return a + b

class TestMath(unittest.TestCase):          # All unittest test cases must be inside a class that inherits from unittest.TestCase
    def test_add(self):                     # Any method starting with test_ will be run automatically by unittest.
        self.assertEqual(add(2, 3), 5)      # self.assertEqual(x, y) checks whether: add(2,3) is equal to rightside value 5 

suite = unittest.TestLoader().loadTestsFromTestCase(TestMath)
unittest.TextTestRunner().run(suite)
```
