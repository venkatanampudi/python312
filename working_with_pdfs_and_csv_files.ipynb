### TASK : Download the PDF from Google Drive link and find the phone number that is in the document.

```python
import PyPDF2
```

```python
f = open('Find_the_Phone_Number.pdf','rb')
```

```python
pdf = PyPDF2.PdfReader(f)
```

```python
len(pdf.pages)
```

```python
import re
```

```python
pattern = r'\d{3}.\d{3}.\d{4}'
```

```python
all_text = ''

for n in range (len(pdf.pages)):
    page = pdf.pages[n]
    page_text = page.extract_text()

    all_text = all_text+' '+page_text
```

```python
#all_text
```

```python
re.search(pattern,all_text) # this only gives the first match ...which is not useful for us
```

```python
re.findall(pattern,all_text)
```

```python
for match in re.finditer(pattern,all_text):
    print(match)
```
