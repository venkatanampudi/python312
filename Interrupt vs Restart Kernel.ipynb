```python
### Difference between Interrupt Kernel vs Restart Kernel

##### Interrupt Kernel: Stop this cell, but do not erase what I've done so far
* It stops the current execution of the cell
* Use when cell execution taking long time 
* This mode keeps all the variables stored in memory
* This mode keeps all the imported libraries and previous cell results

##### Restart Kernel : Start from zero and run the notebook again (its like closing python and opening a fresh session)
* Completely reset the Python environment
* This mode flushes - all variables, imports, stored results, loaded data in memory
```
