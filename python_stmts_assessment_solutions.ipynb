### 2. Use range() to print all the even numbers from 0 to 10 

```python
# Method1 - using for loop with condition
for num in range(0,11):
    if num%2 == 0:
        print(num)
```

```python
# Method2 - direct list 
list(range(0,11,2))
```

```python
# Method3 - List comprehensions
[num for num in range(0,11) if num%2==0]
```

### 3. Use a list comprehension to create a list of all numbers between 1 and 50 that are divisible by 3

```python
[num for num in range(1,51) if num%3 == 0]
```

### 5. Write a program that prints the integers form 1 to 100. But for multiples of three print "Fizz" instead of the number and for the multiples of five print "Buzz". For numbers which are multiples of both three and five print "FizzBuzz"


```python
# check first with multiplese of three and five then with 3 then with 5. If we do first 3, next 5 then 3&5 get diff results.
for num in range(1,101):
    if num%3 == 0 and num%5 == 0:
        print('FizzBuzz')
    elif num%3 == 0:
        print('Fizz')
    elif num%5 == 0:
        print('Buzz')
    else:
        print(num)
    
```

```python
for num in range(1,101):
    if num%3 == 0:
        print('Fizz')
    elif num%5 == 0:
        print('Buzz')
    elif num%5 == 0 and num%5 == 0:
        print('FizzBuzz')
    else:
        print(num)
```

```python
for num in range(1,21):
    if num%2 == 0:
        print('Fizz')
    elif num%4 == 0:
        print('Buzz')
    elif num%2 == 0 and num%4 == 0:
        print('FizzBuzz')
    else:
        print(num)
```
