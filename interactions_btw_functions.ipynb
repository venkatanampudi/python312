```python
# create own function 
def shuffle_list(mylist):
    shuffle(mylist)
    return mylist # this can be assigned to variable and use it later
```

```python
result = shuffle_list(ex_list )
```

```python
# Next, define a function that can take a player's guess 
def player_guess():

    guess=''
    while guess not in ['0','1','2']:
        # while guess not in this input, ask for to enter either 0 or 1 or 2
        guess=input('Pick a number: 0,1, or 2')

    return int(guess)
    
```

```python
def check_guess(mylist,guess): # passing list and player's guess to this function
    if mylist[guess] == 'O':
        print("Correct")
    else:
        print("Wrong guess !")
        print(mylist)
```

```python
# 1. Initial list 
mylist = ['','O','']

# 2. Shuffle the list 
mixedup_list = shuffle_list(mylist)

# 3. User guess
guess = player_guess()

# 4. Check the guess 
check_guess(mixedup_list,guess)
```
