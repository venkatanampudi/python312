#### seed() 
#### random.seed(value) sets the starting point for random number generation. If you use the same seed, Python will produce the **same sequence** of random numbers.
* calling seed() with same value reproduces same behavior
* If you don't set seed , Python uses system time to reproduce the results so results change everytime.
* Seed does NOT make Python more random, it makes it less random (predictible)
* seed(0),seed(42),seed(123) - common choices . ML community often uses seed(42) (answer to life)
#### Why seed() is imp in Data Science/ML? 
* reproduces experiments, test/test splits
* reproduces model training randomness
* compare algorithms fairly
* debug consistent results

```python
random.seed(101)

random.randint(0,100)

```
