* int --> whole numbers -->1, 5, 10
* float --> numbers with decimal points --> 10.9, 2.3, 100.98
* str --> __ordered__ sequence of characters --> "Hello", "Venkat"
* list --> __ordered__ sequence of objects --> [10,"hello",200.5]
* dict --> __unordered__ key:value pairs --> {"key1":"value1","key2":"value2"}
* tup --> __ordered immutable__ sequence of objects --> (10,"hello", 100.10)
* set --> __unordered__ collection of unique objects --> "{"a","b"}
* bool --> logical value indicating __True or False__
