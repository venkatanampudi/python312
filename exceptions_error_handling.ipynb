#### We can use error handling to let the script continue with other code, even if there is an error.
##### *try:* This is the block of code to be attempted (may lead to an error)
##### *except:* Block of code will execute in case there is an error in *try* block 
##### *finally:* A final block of code to be executed, regardless of an error


```python
def add(n1,n2):
    print(n1+n2)
```

```python
number1 = 10 
```

```python
number2 = input("Please provide a number: ")
```

```python
add(number1,number2)
```

```python
try:
    # WANT TO ATTEMPT THIS CODE and MAY HAVE AN ERROR 
    result = 10+10

    # except block execute incase if there is an error in try block.
    # Here no error in try block so no TypeError
except:
    print("Hey it looks like you aren't adding correctly!")
```

```python
#INDUCE AN ERROR IN try BLOCK
try:
    # WANT TO ATTEMPT THIS CODE and MAY HAVE AN ERROR 
    result = 10 + '10'

    # except block execute incase if there is an error in try block.
    # Here no error in try block so no TypeError
except:
    print("Hey it looks like you aren't adding correctly!")
```

```python
#WITH NO ERROR IN try BLOCK

try:
# WANT TO ATTEMPT THIS CODE and MAY HAVE AN ERROR 
    result = 10 + 10


# except block execute incase if there is an error in try block.
# Here no error in try block so no TypeError
except:
    print("Hey it looks like you aren't adding correctly!")

# IF THERE IS NO ERROR IN try BLOCK, JUMP TO else BLOCK TO CONTINUE
else:
    print("Add went well !")
    print(result)
```

```python
try:
    f =  open('testfile','r')
    f.write("Write a test line")

except TypeError:
    print("There is a type error")
except OSError:
    print("Hey you have an OS Error")
    
finally:
    print("I alway run")
```

```python
# USING WHILE LOOP TO CONTINUALLY ASK USER TO ENTER INT 

def ask_for_int():
    while True:        # KEEP ASKING THE USER TO PROVIDE A INT NUMBER IF HE ENTERS A NON-INT INPUT
        try:
            result = int(input("Plz provide a number: "))
        except:
            print("Whoops... This is not a number")
            continue  # CONTINUE WITH NEXT BLOCK OF CODE
        else:         # MEANING, THERE IS NO EXCEPTION MEANS USER PROVIDED A INT NUMBER 
            print("Yes,thank you")
            break     # BREAK OUT OF THIS WHILE LOOP 
        finally:
            print("End of try/except/finally")
            print("I will always run at the end")
```

```python
# ANOTHER METHOD USING SOME VARIABLE
def ask():

    # Waiting for correct response 
    waiting = True
    while waiting:
        try:
            n = int(input("Please enter a integer number: "))
            print("Your number squared is: ")
            print(n**2)      
        except:
            print("Please enter only a integer number: ")
            continue

        else:
            waiting = False
```

```python
ask()
```
