```python
f = open('fileone.txt','w+')
f.write('ONE FILE')
f.close()
```

```python
f = open('filetwo.txt','w+')
f.write('TWO FILE')
f.close()
```

```python
import zipfile 
```

```python
# CREATING A ZIP FILE
comp_file = zipfile.ZipFile('comp_file.zip','w')
```

```python
# ADDING fileone.txt TO comp_file(zipfile)
comp_file.write('fileone.txt',compress_type=zipfile.ZIP_DEFLATED)
```

```python
# ADDING filetwo.txt TO comp_file(zipfile)
comp_file.write('filetwo.txt',compress_type=zipfile.ZIP_DEFLATED)
```

```python
comp_file.close()
```

```python
zip_obj = zipfile.ZipFile('comp_file.zip','r') #zipfile is library
```

```python
zip_obj.extractall('extracted_content')
```
