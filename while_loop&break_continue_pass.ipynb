# While loops
* while loops will continue to execute a block of code while some condition remains __True__
* For Ex: while my pool is not full, keep filling my pool with water

Syntax: While loop 
while some_boolean_condition:
    # do something
else:
    # do something else

```python
x = 0 
while x < 5:
    print(f'The current value of x is {x}')
    x = x + 1
```

```python
x = 0 
while x < 5:
    print(f'The current value of x is {x}')
    x += 1 # x = x + 1
```

```python
x = 0 
while x < 5:
    print(f'The current value of x is {x}')
    x += 1 # x = x + 1
else:
    print('X is not less than 5')
```

```python
x = 50 
while x < 5:
    print(f'The current value of x is {x}')
    x += 1 # x = x + 1
else:
    print('X is not less than 5')
```

# break, continue, pass 
* We can break, continue, and pass statements in our loops to add additional functionality for various cases. The 3 stmts are defined by:
- break: Breaks out of the current closest enclosing loop.
- continue: Goes to the top of the closest loop
- pass: Does nothing at all 

```python
x = [1,2,3]

for item in x:
    # comment 
```

```python
x = [1,2,3]
for item in x:
    # comment
    pass # pass doesn't do anything these. 

print('End of my script')
```

```python
mystring = 'Hello'

for letter in mystring:
    print(letter)

```

```python
mystring = 'Hello'
# continue will say go back to loop and continue from there 
for letter in mystring:
    if letter == 'o':
        continue 
    print(letter)
```

```python
mystring = 'Python'
for letter in 'Python':
    if letter == 'h':
        continue
    print(letter)
```

```python
# breaks the loop from there when condition met.
mystring = 'Python'
for letter in 'Python':
    if letter == 'h':
        break
    print(letter)
```

```python
for letter in 'Getlost':
    if letter == 'l':
        break
    print(letter)
```

```python
# break with while stmt
x=0
while x < 5:
    print(x)
    x += 1
```

```python
x=0
while x < 5:

    if x == 2:
        break
    print(x)
    x += 1
```
