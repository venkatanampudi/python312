# To put everything together, let's create a simple interactive program:
#### This program will:
* Display a list 
* Have a user choose an index position and an input value
* Replace value at index position with user's chosen input value 

```python
game_list = [0,1,2]
```

```python
def display_game(game_list):
    print("Here is the current list: {}".format(game_list))
    #print(game_list)
```

```python
display_game(game_list)
```

```python
def position_choice():

    choice = 'wrong'

    while choice not in ['0','1','2']:
        choice = input("Pick a position (0,1,2): ")

        if choice not in ['0','1','2']:
            print("Sorry, invalid choice ! ")


    return int(choice)
                       
```

```python
position_choice()
```

```python
def replacement_choice(game_list,position):

    user_placement = input("Type a string to place at position: ")

    game_list[position] = user_placement

    return game_list
    
```

```python
replacement_choice(game_list,2)
```

```python
def gameon_choice():

    choice = 'wrong'

    while choice not in ['Y','N']:
        choice = input("Keep playing? (Y or N) ")

        if choice not in ['Y','N']:
            print("Sorry, I dont understand, please choose Y or N ")

    if choice == 'Y':
        return True
    else:
        return False
   
```

```python
gameon_choice()
```

```python
game_on = True
game_list = ['0','1','2']

while game_on:

    display_game(game_list)

    position = position_choice()

    game_list = replacement_choice(game_list,position)

    display_game(game_list)

    game_on = gameon_choice()
```
