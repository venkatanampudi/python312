```python
# INTERSECTION CREATES A NEW SET AND DO NOT CHANGE THE ORIGINAL SET

A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

C = A.intersection(B)
print(C)   # {3, 4}
print(A)   # {1, 2, 3, 4}  (unchanged)
```

```python
# INTERSECTION_UPDATE UPDATES THE EXISTING SET AND RETURNS NONE
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

C = A.intersection_update(B)
print(C)
print(A)
print(B)
```
