# Python comparison Operators - Booleans
- Here, a,b are operands and "== or != or similar" are operators
| Operator  | Desc |Example |
|-------|-----|---------|
| ==    | If the values of two operands are equal,then the condition becomes true          | (a==b) is not true |
| !=    | If the values of two operands are not equal,then the condition becomes true      | (a!=b) is true     | 
| <>    | If the values of two operands are not equal,then the condition becomes true (deprecated in 3.x)      | (a<>b) is true. This similar to != operator         | 
| a>b    | If the values of left operand is greater than the value of right operand,then the condition becomes true  | (a>b) is not true     |
| a<b    | If the values of left operand is less than the value of right operand,then the condition becomes true  | (a<b) is not true     |
| a>=b    | If the values of left operand  is greater than or equal to the value of right operand,then the condition becomes true|(a>=b) is not true|
| a<=b    | If the values of left operand is lesser than or equal to the value of right operand,then the condition becomes true|(a>=b) is true|

```python
'2' == 2
```

```python
3.0 == 3
```

```python
2 != 2
```

```python
2 <> 3 # THIS OPERATOR DEPRECATED IN PYTHON 3.x
```

# Chaining comparison Operators - Logical
#### We can use logical operators to combine comparisons: 
* and 
* or 
* not 


```python
1 < 2 < 3
```

```python
1 < 2 and 2 > 3 
```

```python
'h' == 'h' and 2 == 2
```

```python
('h' == 'h') and (2 == 2) # with parenthesis is better for readability purposes
```

```python
 100 == 1 or 2 == 2 
```

```python
1==2 or 2>1
```

```python
1==1 or 100==1
```

```python
1==1 and 100==100
```
