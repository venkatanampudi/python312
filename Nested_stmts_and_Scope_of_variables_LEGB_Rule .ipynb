# Nested statements and Scope of variables : LEGB Rule 
##### L : __Local__ --> Names assigned in any way within a function (def or lambda) and not declared global in that function 
##### E : __Enclosing function locals__ --> Names in the local scope of any and all enclosing functions (def or lambda), from inner to outer.
##### G : __Global (module)__ --> Names assigned at the top-level of a module file, or declared global in a def within the file 
##### B : __Built-in (Python)__ --> Names preassigned in the built-in names module: open, range, Syntax Error...
* only above Global level is Built-in function

```python
# GLOBAL
name = 'IM A GLOBAL VARIABLE'

def greet():

    #ENCLOSING
    name = 'IM ENCLOSING VARIABLE' #if we comment this, it will search for higher level which is global variable (name = 'THIS IS A GLOBAL STRING') and return the result with this variable
    
    def hello():
        #LOCAL 
        name = 'IM A LOCAL'
        print(name) # this executes first as it is LOCAL and since "name" is not found local it will jump to Enclosing function greet()

    hello() 


    
```

```python
greet()
```

```python
# THIS IS TO EXPLAIN LOCAL VARIABLES and GLOBAL KEYWORDS 
x = 50

def func(x):
    print(f'X is {x}')
    #LOCAL REASSIGNMENT
    x = 200 # this x is only scope to inside this func(x)
    print(f'I just locally changed X to {x}')
```

```python
func(x)
```

```python
print(x) # this print 50 because 200 is assigned inside local func and it is not going to effect at higher scope
```

```python
# Lets we want assing this global x reassign in local func(x)
x = 50

def func():
    global x
    print(f'X is {x}')
    #LOCAL REASSIGNMENT ON A GLOABAL VARIABLE
    x = 'NEW VALUE' # this x is only scope to inside this func(x)
    print(f'I just locally changed global X to {x}')
```

```python
func()
```

```python
print(x)
```

```python
func()
```

```python
print(x)
```

```python
# Instead of using "global" keyword (which is quite dangerous in larger code), better to pass the global value to func and return x
# Basically it is like this : x=50 passing to func(x) and print then reassign the same x to NEW VALUE then print

x = 50

def func(x):
    
    print(f'X is {x}')
    #LOCAL REASSIGNMENT ON A GLOBAL VARIABLE
    x = 'NEW VALUE' # this x is only scope to inside this func(x)
    print(f'I just locally changed global X to {x}')
    return x
```

```python
print(x)
```

```python
x = func(x) # this is clear reassignment here. Specifically grab that global x and reassign the output of the func(x)
```

```python
print(x) # now this x is assign with 'NEW VALUE"
```
