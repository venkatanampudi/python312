### Print Formatting with Strings
- Often you will want to **inject** a variable into a string for printing. For ex: my_name="Venkat" print("Hello" + my_name)
- There are multiple ways to format string for printing variable in them
- This is known as __string interpolation.__
- Two methods:
  - .format() method
  - f-strings (formatted string literals)
- Formatting with the .format() method - A good way to format objects into strings for print statements is with the string .format() method. The syntax is : 'String here{} then also {}'.format('something1','something2')

```python
# The syntax is : "String here{} then also {}".format("something1","something2")
# strings can be inserted by index position
```

```python
print('This is a string{}'.format(' INSERTED'))
```

```python
print('The {} {} {}'.format('fox','brown','quick'))
```

```python
# Lets say we want The quick brown fox, pass thoese indexes into {}
print('The {2} {1} {0}'.format('fox','brown','quick'))
```

```python
print('The {0} {0} {0}'.format('fox','brown','quick'))
```

```python
# we can also assign variables and insert those
print('The {q} {b} {f}'.format(f='fox',b='brown',q='quick'))
```

### Float formatting follows "{value:width.precision f}"
* width --> total printed size
* precision --> number of digits after decimal

```python
result=100/7
```

```python
result
```

```python
print("The result is {}".format(result))
```

```python
print("The result is {r}".format(r=result))
```

```python
print("The result is {r:10.3f}".format(r=result))
```

```python
print("The amount is {r:10.5f}".format(r=result))
```

```python
print("The amount is {r:1.3f}".format(r=result))
```
