### Problem 1: Fill in the Line class methods to accept coordinates as a pair of tuples and return the slope & distance of the line

```python
class Line:

    def __init__(self,coor1,coor2): # passing coordinates as tuples
        self.coor1 = coor1
        self.coor2 = coor2

    def distance(self):
        x1,y1 = self.coor1
        x2,y2 = self.coor2 
        return ((x2-x1)**2 + (y2-y1)**2)**0.5    #distance formula: d = sq.rt(sq(x2-x1)+sq(y2-y1)). In Python, multiply with 0.5 for sq.root

    def slope(self):                             # slope formula: m = (y2-y1) / (x2-x1)
        x1,y1 = self.coor1
        x2,y2 = self.coor2
        return (y2-y1) / (x2-x1)
```

```python
# passing coordinates
c1 = (3,2)
c2 = (8,10)
```

```python
# create myline object with class "Line" reference and pass c1 & c2 
myline = Line(c1,c2)
```

```python
myline.distance()
```

```python
myline.slope()
```

### Problem 2: Find the volume and surface area of the cylinder with given height and radius

```python
# Cylinder class

class Cylinder:

    def __init__(self,height=1,radius=1):
        self.height = height
        self.radius = radius 

    def volume(self):
        return self.height * 3.14 * (self.radius)**2

    def surface_area(self):
        top = 3.14 * (self.radius**2)
        return (2*top) + (2*3.14*self.radius*self.height)
```

```python
mycyl = Cylinder(2,3)
```

```python
mycyl.volume()
```

```python
mycyl.surface_area()
```

```python
# Some simple class to show the method's value to the current value 

class Simple():

    def __init__(self,value):
        self.value = value 

    def add_to_value(self,amount):    # this method do not return anything but it takes init method's value and add to the amount 
        self.value = self.value + amount 
```

```python
class Account():

    def __init__(self,owner,balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self,dep_amt):
        self.balance = self.balance + dep_amt     # also write like self.balance += dep_amt
        print(f"Added {dep_amt} to the balance") 

    def withdrawl(self,wd_amt):
        if self.balance >= wd_amt:  # checking the balance amount before withdrawl 
            self.balance = self.balance - wd_amt    # also write like self.balance -= wd_amt
            print("Withdrawl accepted")
        else:
            print("Sorry, not enough funds available to withdraw!!!")

    def __str__(self):
        return f"Owner: {self.owner} \nBalance: {self.balance}"
    
```
