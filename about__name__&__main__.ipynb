# Its about  `__name__` & `__main__`
* if `__name__` == "`__main__`" 

#### Sometimes when you are importing from a module, you would like to know whether a modules function is being using as an import, or if you are using the original.py file of that moduel

#### When we run python myscript.py, all of the code at the indentation level 0 is going to execute. That means all the print stmts, all the functions defined at indentation level 0 execute. Unlike in other languages,in Python there is no main() function to call some where in your script to run automatically. 

# Built-in variable `__name__` gets assinged the string  `__main__` (`__name__` == "`__main__`") in the background when you run the .py file directly. 

```python
# one.py
def func():
    print("FUNCTION IN ONE.PY")


print("TOP LEVEL IN ONE.PY")

if __name__ == "__main__":
    print("ONE.PY is being run directly")
else:
    print("ONE.PY has been imported")
```

```python
# two.py
import one 
print("TOP LEVEL IN TWO.Py")

one.func()

if __name__ == "__main__":
    print("TWO.PY is being run directly")
else:
    print("TWO.PY has been imported")
```

```python
# in general we see this. Define functions first and execution logic in __main__
def func():
    print("FUNCTION IN ONE.PY")

def func1():
    pass

def func2():
    pass
    
if __name__ == "__main__":
    # RUN THE SCRIPT 
    func()
    func1()
    func2()

    
```
