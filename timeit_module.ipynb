### TIMING YOUR PYTHON CODE
- As you go on with Python, you will discover multiple solutions for a single task and you may find yourself trying to figure out the most efficient approach.
- An easy way to do this is to time your code's performance
- 3 ways to do this:
- ``` Simply tracking time elapsed```
- ``` Using the timeit module```
- ``` Special %%timeit "magic" for Jupyter Notebook```

````python
def func_one(n):
    return [str (num)for num in range (n)]
````

````python
func_one(10)
````

````python
def func_two(n):
    return list(map(str,range(n)))
````

````python
func_two(10)
````

````python
import time
````

````python
# Current time before run
start_time = time.time()
# run the code 
result = func_one(100000)
# current time after running the code
end_time = time.time()
# ELAPSED TIME
elapsed_time = end_time - start_time
print(elapsed_time)
````

````python
# Current time before run
start_time = time.time()
# run the code 
result = func_two(100000)
# current time after running the code
end_time = time.time()
# ELAPSED TIME
elapsed_time = end_time - start_time
print(elapsed_time)
````

````python
# we cannot compare these 2 times becoz they ran so fast
# func_one is faster but we cannot say how much faster it is compare to func_two
# use timeit module
````

````python
import timeit 
````

````python
stmt = '''
func_one(100)
'''
````

````python
setup = '''
def func_one(n):
    return [str (num)for num in range (n)]
'''
````

````python
timeit.timeit(stmt,setup,number=10000000)
````

````python
stmt2 = '''
func_two(100)
'''
````

````python
setup2 = '''
def func_two(n):
    return list(map(str,range(n)))
'''
````

````python
timeit.timeit(stmt,setup,number=10000000)
````

````python
%%timeit 
func_one(100)
````

````python
%%timeit
func_two(100)
````
