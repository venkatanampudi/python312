#### 3. Write a Python function that accepts a string and calculates the number of upper case letters and lower case letters
#### Sample string "Hello Mr.Rogers, how are you this fine Tuesday?"
#### Output: Upper case chars = 4, lower case chars = 33
#### HINT: Two string methods that might be useful: .isupper() and .islower()

```python
# METHOD 1 - using for loop
def up_low(s):
    lowercase = 0
    uppercase = 0 

    for char in s:
        if char.isupper():
            uppercase += 1
        elif char.islower():
            lowercase += 1
        else: 
            pass # if string contains "?"

    print(f'Original String: {s}')
    print(f'Lowercase count: {lowercase}')
    print(f'Uppercase count: {uppercase}')
    
```

```python
s = "Hello Mr.Rogers, how are you this fine Tuesday?"
up_low(s)
```

```python
s = "Hey...Mr.Venkat, How are you doing today?"
up_low(s)
```

```python
# METHOD 2 - using dictionaries 
def up_low(s):
    d = {'upper':0,'lower':0} 

    for char in s:
        if char.isupper():
            d['upper'] += 1
        elif char.islower():
            d['lower'] += 1
        else: 
            pass # if string contains "?"

    print(f'Original String: {s}')
    print(f'Lowercase count: {d["lower"]}')
    print(f'Uppercase count: {d["upper"]}')
```

```python
s = "Hey...Mr.Venkat, How are you doing today?"
up_low(s)
```

#### 4. Write a Python function that takes a list and returns a new list with unique elements of the first list
#### Sample list : [1,1,1,1,2,2,3,3,3,3,3,4,5]
#### Unique list: [1,2,3,4,5]

```python
# Using set() and it is most Pythonic way
def unique_list(sample_list):
    return list(set(sample_list))
```

```python
unique_list([1,1,1,1,2,2,3,3,3,3,3,4,5])
```

```python
# Method 2 - manually
def unique_list1(sample_list):
    seen_numbers = []
    for num in sample_list:
        if num not in seen_numbers:
            seen_numbers.append(num)
    return seen_numbers
            
```

```python
unique_list1([1,1,1,1,2,2,3,3,3,3,3,4,5])
```

#### 5. Write a Python function to multiply all the numbers in a list.
#### sample list : [1,2,3,-4]
#### Expected output : -24

```python
def multiply(numbers):
    total = 1 # IF ASK IS TOTAL, THIS TOTAL MUST START WITH "0". SINCE THIS IS MULTIPLY, TOTAL=1 SO ANY NUMBER MULTIPLY WITH 1 IS THE SAME NUMBER
    for num in numbers:
        total = total * num
    return total
```

```python
multiply([1,2,3,-4])
```

```python
multiply([100,100])
```

```python
multiply([0,100])
```

#### 6. Write a Python function that checks whether a word or phrase is a palindrome (reads the same backward as forward) or not. Must return True or False
#### Ex1: madam --> madam, kayak --> kayak, racecar --> racecar, nurses run --> nurses run, zilliz --> zilliz 
#### U can use .replace() to deal with spaces in the string

```python
def palindrome(s):
    # REMOVE SPACES STRING
    s = s.replace(' ','')
    # Check if string is == reversed version of the string 
    return s
```

```python
palindrome('nurses run')
```

```python
def palindrome(s):
    # REMOVE SPACES STRING
    s = s.replace(' ','')
    # Check if string is == reversed version of the string 
    return s == s[::-1] # this means - all the way from the begining to ending with step size -1 (means grab from last char of the given string)
    
```

```python
palindrome('madam')
```

```python
palindrome('nurses run')
```

```python
palindrome('m a d a m')
```

```python
palindrome('zillow')
```

```python
palindrome('zilliz')
```

#### 7. Write a Python function to check whether a string is pangram or not. (Assume the string passed in does not have any punctuation)
#### Pangram are words or sentences containing every letter of the alphabet at least once 
#### Ex: "The quick brown fox jumps over the lazy dog"
#### You may want to use .replace() method to get rid of spaces

```python
import string 

def ispangram(str1,alphabet= string.ascii_lowercase):

    # create a set of the alphabet
    alphaset = set(alphabet)
    #print(alphaset)

    # Remove any spaces from the input string
    str1 = str1.replace(' ','')

    # Convert into lower case
    str1 = str1.lower()

    # Grab all unique letters from the string with set() method
    str1 = set(str1)
    
    # compare alphabet set == string set input
    return str1 == alphaset
    
```

```python
ispangram("The quick brown fox jumps over the lazy dog")
```
