```python
class Animal():# THIS IS BASE CLASS

    def __init__(self): # __init__ -->It initializes (sets up) the object when a class is created and It runs automatically whenever you create an instance of the class.
        print("ANIMAL CREATED")

    def who_am_i(self):
        print("I AM AN ANIMAL")

    def eat(self):
        print("I AM EATING")


    
```

```python
myanimal = Animal()
```

```python
myanimal.eat()
```

```python
myanimal.who_am_i()
```

```python
class Dog(Animal): # THIS IS DERIVED CLASS AS IT IS INHERITED FROM THE Animal class
    def __init__(self):
        Animal.__init__(self)
        print("Dog created")

    # OVERWRITING ONE OF THE OLDER METHODS OF Animal class
    def who_am_i(self):
        print("I am a dog!")
```

```python
mydog = Dog()
```

```python
mydog.who_am_i()
```

```python
# WE CAN ALSO ADD NEW METHODS 

class Dog(Animal):
    def __init__(self):
        Animal.__init__(self)
        print("Dog created")
    
    def eat(self): # overwriting with same method name
        print("I am a dog and eating")
    
    def bark(self): # adding a new method
        print("WHOOF")
```

```python
mydog = Dog()
```

```python
mydog.bark()
```

```python
mydog.eat()
```

```python
# POLYMORPHISM
class Dog(): 
    
    def __init__(self,name):
        self.name = name

    def speak(self):
        return self.name + " says woof!"
    
```

```python
class Cat(): 
    
    def __init__(self,name):
        self.name = name

    def speak(self):
        return self.name + " says meow!"
```

```python
teddy = Dog("Teddy")
felix = Cat("Felix")
```

```python
print(teddy.speak())
```

```python
print(felix.speak())
```
