#### 11. Lesser of two evens. Write a function that returns the lesser of two given numbers if both numbers are even,but returns the greater if one or both numbers are odd.
#### lesser_of_two_evens(2,4) --> 2 
#### lesser_of_two_evens(2,5) --> 5

```python
# METHOD 1
def lesser_of_two_evens(a,b):
      
    if a%2 == 0 and b%2 == 0:
        # BOTH NUMBERS ARE EVEN !
        if a < b:
            result = a
        else:
            result = b 
    else:
        # ONE OR BOTH NUMBERS ARE ODD ! 
        if a > b:
            result = a
        else:
            result = b
    return result
```

#### 14. OLD MACDONALD : Write a function that capitalizes the first and fourth letters of a name
#### old_macdonald('macdonald') --> MacDonald
#### Note: 'macdonald'.captitalize() returns 'Macdonald'

```python
# METHOD 1 (with upper method)

def old_macdonald(name):
    first_letter = name[0]
    inbetween = name[1:3]
    fourth_letter = name[3]
    rest = name[4:]

    return first_letter.upper() + inbetween + fourth_letter.upper() + rest

   
```

```python
old_macdonald('macdonald')
```

```python
# METHOD 2 with capitalize method

def old_macdonald(name):
    first_half = name[:3]
    second_half = name[3:]
      
    return first_half.capitalize() + second_half.capitalize()
```

#### 15. MASTER YODA: Given a sentence, return a sentence with the words reversed. 
#### master_yoda('I am home') --> 'home am I'
#### master_yoda('We are ready') --> 'ready are we'

```python
def master_yoda(text):
    wordlist = text.split()
    #print(wordlist)
    reverse_wordlist = wordlist[::-1] # -1 from reverse order of words
    # print(reverse_wordlist)
    return ' '.join(reverse_wordlist)
```

```python
master_yoda('We are ready')
```

##### blackjack(5,6,7) --> 18

##### blackjack(9,9,11) --> 19

```python
def blackjack(a,b,c):
    if sum([a,b,c]) <= 21:
        return sum([a,b,c])
    elif 11 in [a,b,c] and sum([a,b,c]) <=31:
        return sum([a,b,c])-10
    else:
        return 'BUST'
```

#### 19.SUMMER OF '69: Return the sum of the numbers in the array, except ignore sections of numbers starting with a 6 and extending to the next 9 (every 6 will be followed by at least one 9). Return 0 for no numbers.
##### summer_69([1,3,5]) ---> 9
##### summer_69([4,5,6,7,8,9]) ---> 9
##### summer_69([2,1,6,9,11]) ---> 14 


```python
def summer_69(arr):
    total = 0
    add = True

    for num in arr:
        if add:
            if num != 6:
                total += num
            else:
                add = False  # start ignoring numbers
        else:
            if num == 9:
                add = True  # stop ignoring numbers

    return total
```

```python
summer_69([4,5,6,7,8,9])
```

```python
summer_69([2,1,6,9,11])
```

#### 20. SPY GAME: Write a function that takes in a list of intergers and returns True if it contains 007 in order.
#### spy_game([1,2,1,0,0,7,5]) ==> True
#### spy_game([1,0,2,4,0,7,7]) ==> True (00 before 7 here)
#### spy_game([1,7,2,0,4,5,0]) ==> False

```python
def spy_game(nums): 
   
    code = [0,0,7,'x']
    # [0,7,'x']
    # [7,'x']
    # ['x'] length = 1
    for num in nums:
        if num == code[0]:
            code.pop(0)
    return len(code) == 1
```

```python
spy_game([1,2,1,0,0,7,5])
```

```python
spy_game([1,7,2,0,4,5,0])
```

```python
spy_game([0,0,7])
```

```python
def count_primes(num):
    # check for 0 or 1 input 
    if num < 2:
        return 0 

    #####################
    # 2 or greater
    ####################

    # Store our prime numbers 
    primes = [2]

    # counter going upto the input num
    x= 3
    
    # x is going thru every number upto input num

    while x <= num:
        # check if x is prime
        for y in range(3,x,2): #because from 3, every second number is even
            if x%y == 0:
                x += 2
                break
        else:
            primes.append(x)
            x += 2
    print(primes)
    return len(primes)
```

#### HINT: Consider making a dictionary of possible patterns, and mapping the alphabet to specific 5-line combinations of patterns.
