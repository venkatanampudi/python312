# Useful Operators in Python 
* __range()__ function
* __enumerate()__ --> enumerate() is a built-in function that adds a __counter (index) to an iterable__ (like a list, tuple, string) while looping.
* __zip()__ function --> __to combine (or zip together)__ elements from two or more iterables (like lists, tuples) position-by-position
* __in__ operator --> to __check__ if variable value in the iterable object and return boolean
* __input()__ function --> to input something or user input. This always expects string.
 ---
- The term __inplace in Python refers to an operation that changes the original object directly, instead of creating a new one.__

```python
# without enumerate() function
index_count = 0
for letter in 'abcde':
    print('At index {} the letter is {}'.format(index_count,letter))
    index_count += 1
```

```python
# enumerate() built-in function 
# enumerate takes any iterable object (list,tuple,string) and adds a counter (or index) while looping
word = 'abcde'
for item in enumerate(word):
    print(item)
```

```python
# index starts from "0" by default. 
for index,letter in enumerate(word):
    print(index,letter)
    
```

```python
# You can pass "start" to bypass default index if needed.
for index,letter in enumerate(word,start=1): 
    print(index,letter)
```

```python
# You can pass "start" to bypass default index if needed.Here, start index is 5
nums = [10,20,30]

for i,n in enumerate(nums,start=5):
    print(i,n)
```

```python
mylist1 = [1,2,3]
mylist2 = ['a','b','c']

# zip(mylist1,mylist2)
# Here, zip function wont return anything since it is inplace function.
zipping = zip(mylist1,mylist2)
```

```python
zipping
```

```python
# zip function 
for item in zip(mylist1,mylist2):  
    print(item)
```

```python
# zip with 3 lists
mylist1 = [1,2,3]
mylist2 = ['a','b','c']
mylist3 = [100,200,300]

for item in zip(mylist1,mylist2,mylist3):
    print(item)
```

```python
# zip func will go as far as which is the shortest. It wont zip together if list is uneven like below:

mylist1 = [1,2,3,4,5,6]
mylist2 = ['a','b','c','d']
mylist3 = [100,200,300]

for item in zip(mylist1,mylist2,mylist3):
    print(item)

```

```python
# convert the zip into list 
list(zip(mylist1,mylist2,mylist3))
```

```python
for a,b,c in zip(mylist1,mylist2,mylist3):
    print(a,b,c)
```
