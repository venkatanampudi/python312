### Object Oriented Programming 
* Object Oriented Programming (OOPS) allows programmers to create their own objects that have methods and attributes.
* Recall that after defining a string, list, dict, or other objects, you were able to call methods off of them with the .method_name() syntax
* These methods act as functions that use information about the objects, as well the object itself to return results, or change the current object.
* For ex - this includes appending to a list, or counting the occurences of an element in a tuple
* OOP allows users to create their own objects
* In general, OOP allows us to create code that is repeatable and organized

# Class object attributes 

```python
class Dog():

    # CLASS OBJECT ATTRIBUTES AND THESE ARE SAME FOR ANY INSTANCE OF THE CLASS.
    species = 'mammal'
    
    def __init__(self,breed,name):
        
        # most of the times parameter names and attribute names are same.
        self.breed = breed
        self.name = name

        # a method is a function is inside of the class
        # attributes never had (). Attributes something to execute 
        # methods need to be executed 
        
    #Operations/Actions --> Methods . Usually methods take information of the class, for ex "name" here
    #def bark(self):
        #print("Woof!")

    def bark(self):
        print("Woof! My name is {}".format(self.name)) 

    def bark(self,number):
        # Methods also can take outside parameters like number here
        print("Woof! My name is {} and the number is {}".format(self.name,number)) 
```

```python
my_dog = Dog('Lab','Frankie')
```

```python
my_dog.species # this is class object attribute and no ()
```

```python
my_dog.name
```

```python
my_dog.bark() # This is method and should execute with ()
```

```python
my_dog.bark
```

```python
my_dog.bark()
```

```python
my_dog.bark(9)
```

```python
class Circle():

    # Class object attribute
    pi = 3.14 

    def __init__(self,radius=1):

        self.radius = radius 

    # METHOD 
    def get_circumference(self):
        return self.radius * self.pi * 2
        
```

```python
my_circle = Circle() # default radius value 
```

```python
my_circle.pi
```

```python
my_circle = Circle(30) # overwriting the radius value with 30
```

```python
my_circle.radius
```

```python
my_circle.get_circumference()
```

```python
class Circle():

    # Class object attribute
    pi = 3.14 

    def __init__(self,radius=1):

        self.radius = radius 
        #self.area = radius * radius * self.pi  # assigning object attribute directly. Attributes no need to come from constructor object always.
        # PiR2 is area of circle formula
        # Here, self in self.pi references a particular instance of this object or we can reference with Circle.pi as well if it is class object attribute
        self.area = radius * radius * Circle.pi

        
    # METHOD 
    def get_circumference(self):
        #return self.radius * self.pi * 2
        return self.radius * Circle.pi * 2
```

```python
my_circle = Circle(30)
```

```python
my_circle.radius
```

```python
my_circle.area
```

```python
my_circle.get_circumference()
```
