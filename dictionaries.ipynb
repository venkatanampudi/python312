# Dictionaries
* Dictionaries are **unordered mappings** for storing objects.
* __ordereddict__ object can be used for ordering
* Dictionaries use a __KEY-VALUE__ pairs
* KEY-VALUE pairs allows users to quickly grab objects __without needing to know an index location__
* Dictinaries use {} braces and colons to signify the key and their assocated values. {'key1':'value1','key2':'value2'}

### When to choose Dictionaries or Lists ?
* Dictionaries: objects retrieve by key name, unordered and cannot be sorted
* Lists: Objects retrieved by location, ordered sequence can be indexed or sliced

```python
week={'MON':17, 'TUE':18, 'WED':19,'THU':20,'FRI':21}
```

```python
week.keys()
```

```python
week.values()
```

```python
week['THU']
```

```python
# DICT IS NOT BUILT FOR REVERSE LOOKUP.
value_to_find=20
for key,value in week.items():
    if value == value_to_find:
        print(key)
```

```python
week.items()
```

```python
v=21
for key,value in week.items():
    if value == v:
        print(key)
```
