```python
def splicer(mystring):
    if len(mystring)%2 == 0:
        return 'STRING CONTAINS EVEN NUMBER OF CHARS'
    else:
        return 'STRING CONTAINS ODD NUMBER OF CHARS'
```

```python
splicer('ARJUN')
```

```python
splicer('ABCDEFGHIJKLMNOPQRSTUVWXYZ')
```

```python
splicer('PYTHON')
```

```python
splicer('python')
```

```python
names = ['Andy','Eve','Sally']
```

```python
list(map(splicer,names))
```

```python
# filter() - Filters items based on a function that returns True/False
```

```python
#FILTER 
def check_even(num):
    return num%2 ==0
    
```

```python
mynums = [1,2,3,4,5,6]
```

```python
filter(check_even,mynums)
```

```python
list(filter(check_even,mynums))
```

```python

for n in filter(check_even,mynums):
    print(n)
```

```python
# this is normal function
def square(num):
    result = num ** 2
    return result
```

```python
square(10)
```

```python
# converting the above func into lambda expression (lambda func is anonymous and use only once in general)
def square(num): return num ** 2
```

```python
# even without "return"
# before : is input and after : is output
variable1 = lambda num: num ** 2 
```

```python
variable1(25)
```

```python
# lambda with map() function
list(map(lambda num:num**2,mynums))
```

```python
map(lambda num:num**2,mynums)
```

```python
mynums = [1,2,3,4,5,6]
list(filter(lambda num:num%2 ==0,mynums))
```
