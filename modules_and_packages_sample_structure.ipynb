# How to create our own modules and packages ?
#### Modules are just .py scripts that you call another .py script
#### Packages are collection of modules

```python
1. create a folder "python312" in windows C 
2. create a folder "MyMainPackage" in "python312"
2.1 create a mymodule.py & myprogram.py files in "python312"
3. create a subfolder "SubPackage", __init__.py,some_main_script.py files in "MyMainPackage" folder 
4. create __init__.py, mysubscript.py files in "SubPackage"
    
```

```python
# mymodule.py

def my_func():
    print("Hey I am in mymodule.py")
```

```python
#myprogram.py

from MyMainPackage import some_main_script            # importing some_main_script from MyMainPackage 
from MyMainPackage.SubPackage import mysubscript      # importing mysubscript from MyMainPackage.SubPackage
some_main_script.report_main()
mysubscript.sub_report()
```

```python
#some_main_script.py

def report_main():
    print("Hey I am in some_main_script in main package")
```

```python
# mysubscript.py

def sub_report():
    print("Hey I am a function inside mysubscript")
```
